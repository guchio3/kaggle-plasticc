```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_pickle('../lcfit/okumurasan_feats/LCfit_feature_allSN_i_train_v2_20181213.pkl.gz', compression='gzip')
test_df = pd.read_pickle('../lcfit/okumurasan_feats/LCfit_feature_allSN_i_test_v2_20181213.pkl.gz', compression='gzip')
```

```python
def get_okumurasan_feats(lc_df):
    df = lc_df.copy()
    df['dmax_std'] = df[['c42_i_z0_dmax', 'c42_i_z1_dmax', 'c42_i_z2_dmax', 'c42_i_z3_dmax']].std(axis=1)
    df['dmax_mean'] = df[['c42_i_z0_dmax', 'c42_i_z1_dmax', 'c42_i_z2_dmax', 'c42_i_z3_dmax']].mean(axis=1)
    return df
```

```python
train_df = get_okumurasan_feats(train_df)
train_df.columns
```

```python
test_df = get_okumurasan_feats(test_df)
test_df.columns
```

```python
train_df.to_pickle('../lcfit/train_v2_20181213.pkl.gz', compression='gzip')
test_df.to_pickle('../lcfit/test_v2_20181213.pkl.gz', compression='gzip')
```

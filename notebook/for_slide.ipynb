```python
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

def plot_lc(df, target, detected=False):
    if detected:
        df = df.query('detected==1')
    plt.figure(figsize=(8, 5))
    for band in range(6):
        plt_df = df.query(f'passband == {band}')
        #plt.scatter(x=plt_df.mjd, y=plt_df.flux, label=band)
        plt.errorbar(x=plt_df.mjd, y=plt_df.flux, yerr=plt_df.flux_err, label=band, fmt='o')#, color=colors[band], ecolor=colors[band])
    plt.title(f'class: {target}, object_id: {df.object_id.iloc[0]}, detected: {detected}')
    plt.xlim((59800, 60680))    
    plt.xlabel('mjd')
    plt.ylabel('flux')
    plt.legend(['u', 'g', 'r', 'i', 'z', 'y'])
```

```python
target = 90
object_id = trn_meta_df.query(f'target == {target} & ddf== 1').object_id.iloc[6]
plot_lc(trn_set_df.query(f'object_id == {object_id}'), target, detected=False)
```

```python
ae_df = pd.read_pickle('/home/naoya.taguchi/ae_train_flux_v2.pkl.gz', compression='gzip')
```

```python
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']

def plot_lc_ae(df, target, detected=False):
    if detected:
        df = df.query('detected==1')
    df = df.reset_index().rename(columns={'index': 'int_mjd'})
    plt.figure(figsize=(8, 5))
    for band in range(6):
        plt_df = df[[band, 'int_mjd']].rename(columns={band: 'flux'})
        plt_df['flux'] -= plt_df['flux'].mean()
        plt.scatter(x=plt_df.int_mjd, y=plt_df.flux, label=band)
        #plt.errorbar(x=plt_df.int_mjd, y=plt_df.flux, yerr=plt_df.flux_err, label=band, fmt='o')#, color=colors[band], ecolor=colors[band])
    plt.title(f'class: {target}, object_id: {df.object_id.iloc[0]}, detected: {detected}')
    #plt.xlim((59800, 60680))    
    plt.xlabel('compressed mjd')
    plt.ylabel('flux')
    plt.legend(['u', 'g', 'r', 'i', 'z', 'y'])
```

```python
target = 90
object_id = 4389
plot_lc_ae(ae_df.query(f'object_id == {object_id}'), target, detected=False)
```

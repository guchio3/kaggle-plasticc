```python
import pandas as pd
```

```python
g_tr = pd.read_pickle('../lcfit/LCfit_feature_allSN_g_train_v3_20181215.pkl.gz', compression='gzip')
i_tr = pd.read_pickle('../lcfit/LCfit_feature_allSN_i_train_v3_20181215.pkl.gz', compression='gzip')
z_tr = pd.read_pickle('../lcfit/LCfit_feature_allSN_z_train_v3_20181215.pkl.gz', compression='gzip')
r_tr = pd.read_pickle('../lcfit/LCfit_feature_allSN_r_train_v3_20181215.pkl.gz', compression='gzip')
sw_tr =  pd.read_pickle('../lcfit/switch_feature_train_20181202.pkl.gz', compression='gzip')
```

```python
sw_tr
```

```python
oku_train_df = g_tr.merge(i_tr, on='object_id', how='left')
oku_train_df = oku_train_df.merge(z_tr, on='object_id', how='left')
oku_train_df = oku_train_df.merge(r_tr, on='object_id', how='left')
oku_train_df = oku_train_df.merge(sw_tr, on='object_id', how='left')
oku_train_df
```

```python
oku_train_df.to_pickle('../lcfit/the_last_train_okumurasan_feats.pkl.gz', compression='gzip')
```

```python
g_te = pd.read_pickle('../lcfit/LCfit_feature_allSN_g_test_v3_20181215.pkl.gz', compression='gzip')
i_te = pd.read_pickle('../lcfit/LCfit_feature_allSN_i_test_v3_20181215.pkl.gz', compression='gzip')
z_te = pd.read_pickle('../lcfit/LCfit_feature_allSN_z_test_v3_20181215.pkl.gz', compression='gzip')
r_te = pd.read_pickle('../lcfit/LCfit_feature_allSN_r_test_v3_20181215.pkl.gz', compression='gzip')
sw_te =  pd.read_pickle('../lcfit/switch_feature_test_20181202.pkl.gz', compression='gzip')
```

```python
oku_test_df = g_te.merge(i_te, on='object_id', how='left')
oku_test_df = oku_test_df.merge(z_te, on='object_id', how='left')
oku_test_df = oku_test_df.merge(r_te, on='object_id', how='left')
oku_test_df = oku_test_df.merge(sw_te, on='object_id', how='left')
oku_test_df
```

```python
oku_test_df.to_pickle('../lcfit/the_last_test_okumurasan_feats.pkl.gz', compression='gzip')
```
